# KIU Internship Support Survey Analysis

This notebook uses the same functions as `analysis-script.py` so the notebook and command-line workflow stay aligned.

````python
from pathlib import Path
import importlib.util

cwd = Path.cwd()
candidates = [cwd, cwd / 'kiu-internship-support-research', *cwd.parents]
BASE_DIR = next(
    (candidate for candidate in candidates if (candidate / 'analysis' / 'analysis-script.py').exists()),
    None,
)
if BASE_DIR is None:
    raise FileNotFoundError('Could not locate analysis/analysis-script.py from the current notebook folder.')

script_path = BASE_DIR / 'analysis' / 'analysis-script.py'
spec = importlib.util.spec_from_file_location('survey_analysis', script_path)
survey_analysis = importlib.util.module_from_spec(spec)
spec.loader.exec_module(survey_analysis)

df = survey_analysis.load_and_clean_data(survey_analysis.RAW_PATH)
df.head()
````

````python
frequency_df = survey_analysis.categorical_frequency_table(df)
missing_df = survey_analysis.missing_values_table(df)
scale_df = survey_analysis.scale_mean_table(df)
group_df = survey_analysis.readiness_group_comparison(df)
ttest_df = survey_analysis.run_ttest(df)
reliability_df = survey_analysis.reliability_table(df)

frequency_df.head(10)
````
